##### Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sb
import plotly.express as px

titanic = sb.load_dataset('titanic')

fig = px.scatter(titanic, x='age', y='fare', title="Scatter Plot of Age vs Fare in Titanic Dataset", height=500)
fig.show()
```

##### Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px

px.data.tips

fig = px.box(tips, x='day', y='total_bill', title="Box Plot of Total Bill by Day", height=500)
fig.show()
```

##### Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
import plotly.express as px

px.data.tips

fig = px.histogram(tips, x='sex', y='total_bill', pattern_shape='smoker', color='day', height=500)
fig.show()
```

##### Q5. What is Distplot? Using Plotly express, plot a distplot.

A **distplot** (distribution plot) is a graphical representation that combines a histogram and a kernel density estimate to visualize the distribution of a continuous variable in a dataset. It displays the frequency distribution of data points along with an estimated probability density function that represents the underlying continuous distribution.

Plotly Express does not have a direct equivalent of Seaborn's **'distplot'**. However, you can achieve similar functionality using the **'px.histogram()'** function combined with the **'kernel_density'** parameter. Here's how you can plot a distribution plot (distplot) using Plotly Express:

```python
import plotly.express as px

px.data.tips

fig = px.histogram(tips, x='total_bill', marginal='rug', title="Distribution Plot (Distplot) of Total Bill", height=800)
fig.update_layout(bargap=0.1)
fig.show()
```
